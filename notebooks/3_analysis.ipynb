## Institutes

```python
df = pd.read_csv('../outputs/clean/institutes.csv')
```

## Most recent data

```python
# Get the most recent update
df['posti_aggiornati_al'] = pd.to_datetime(df['posti_aggiornati_al'], format='%Y/%m/%d')
df_most_recent = df.loc[df.groupby('id')['posti_aggiornati_al'].idxmax()]
# Adding columns for places available and overcrowding index
df_most_recent['posti_reali'] = df_most_recent['posti_regolamentari'] - df_most_recent['posti_non_disponibili']
df_most_recent['tasso_sovraffollamento'] = round(df_most_recent['posti_occupati'] / df_most_recent['posti_reali'],2)
df_most_recent.head()
```

```python
# Adding institutes' information
# Read institutes' csv

df_info = pd.read_csv('../outputs/raw/institutes_info.csv')
df_info = df_info.rename(columns={'id_istituto': 'id'})

merged_df = pd.merge(df_most_recent, df_info, on='id')

# Adding additional columns
merged_df['posti_reali'] = merged_df['posti_regolamentari'] - merged_df['posti_non_disponibili']

merged_df['tasso_affollamento'] = (merged_df['posti_occupati'] / merged_df['posti_reali']).round(2)

# Keep only relevant columns
df_filtered = merged_df[
  [
    'id',
    'nome',
    'tasso_affollamento',
    'indirizzo',
    'tipo',
   'posti_regolamentari',
    'posti_non_disponibili',
    'posti_reali',
    'posti_occupati',
    'posti_aggiornati_al',
    'personale_polizia_previsti',
    'personale_polizia_effettivi',
    'personale_polizia_aggiornato_a',
    'personale_amministrativi_previsti',
    'personale_amministrativi_effettivi',
    'personale_amministrativo_aggiornato_al',
    'latitudine',
    'longitude'
    ]
    ]

df_filtered.head()
```

```python
df_filtered.head()
```

```python
# Saving csv
df_filtered.to_csv('../outputs/viz/institutes_most_recent.csv', index=False, encoding='UTF-8-sig')
```

## 1. Most recent totals

```python
columns_to_sum = [
    'posti_regolamentari',
    'posti_non_disponibili',
    'posti_reali',
    'posti_occupati',
    'personale_polizia_previsti',
    'personale_polizia_effettivi',
    'personale_amministrativi_previsti',
    'personale_amministrativi_effettivi'
]

# Create a new DataFrame with the sum of the specified columns
df_summed = pd.DataFrame(df_filtered[columns_to_sum].sum()).transpose()

# Optionally, add meaningful names for the new DataFrame
df_summed.index = ['Total Sums']  # Rename the index for clarity

# Display the resulting summed DataFrame
df_summed
```

```python
df_summed['tasso_sovraffollamento'] = (df_summed['posti_occupati'] / df_summed['posti_reali']).round(2)

df_summed.head()
```

```python
df_summed.to_csv('../outputs/viz/institutes_most_recent_totals.csv', index=False)
```

### Institutes - Totals

```python
df = pd.read_csv('../outputs/clean/institutes.csv')
```

```python
df.head()
```

```python
df['posti_aggiornati_al'].value_counts()
```

```python
df.head()
```

```python
# Define the columns you want to sum
columns_to_sum = [
    'posti_regolamentari',
    'posti_non_disponibili',
    'posti_occupati',
    'personale_polizia_previsti',
    'personale_polizia_effettivi',
    'personale_amministrativi_previsti',
    'personale_amministrativi_effettivi'
]

# Group by 'posti_aggiornati_al' and sum the specified columns
aggregated_df = df.groupby('posti_aggiornati_al')[columns_to_sum].sum().reset_index()

# Print or inspect the aggregated DataFrame
aggregated_df
```

```python
aggregated_df.dtypes
```

```python
aggregated_df['posti_reali'] = aggregated_df['posti_regolamentari'] - aggregated_df['posti_non_disponibili']
aggregated_df['indice_sovraffollamento'] = (aggregated_df['posti_occupati'] / aggregated_df['posti_reali']).round(2)

aggregated_df
```

```python
aggregated_df['posti_aggiornati_al'] = pd.to_datetime(aggregated_df['posti_aggiornati_al'])
aggregated_df = aggregated_df.sort_values(by='posti_aggiornati_al')
aggregated_df.reset_index(drop=True, inplace=True)
aggregated_df
```

```python
aggregated_df.to_csv('../outputs/viz/institutes_totals.csv', index=False)
```
