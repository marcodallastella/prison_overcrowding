```python
import pandas as pd
```

```python
df = pd.read_csv('../outputs/clean/istituti_data.csv')
```

```python
df.head()
```

## Most recent

```python
df['posti_aggiornati_al'].value_counts()
```

```python
df['posti_aggiornati_al'] = pd.to_datetime(df['posti_aggiornati_al'], format='%d/%m/%Y')
```

```python
df['posti_aggiornati_al'].max()
```

```python
last_update = df['posti_aggiornati_al'].max()
last_update
```

```python
df_most_recent = df[['id', 'nome', 'tipo', 'posti_regolamentari', 'posti_non_disponibili', 'posti_occupati', 'posti_aggiornati_al']]
df_most_recent
```

```python
df_most_recent = df_most_recent[df_most_recent['posti_aggiornati_al'] == last_update]
df_most_recent
```

```python
df_most_recent['posti_reali'] = df_most_recent['posti_regolamentari'] - df_most_recent['posti_non_disponibili']
df_most_recent['tasso_occupazione'] = round(df_most_recent['posti_occupati'] / df_most_recent['posti_reali'],2)
df_most_recent
```

```python
df_most_recent.head()
```

```python
df_most_recent['posti_aggiornati_al'] = df_most_recent['posti_aggiornati_al'].dt.strftime('%d/%m/%Y')
df_most_recent['posti_aggiornati_al'] = df_most_recent['posti_aggiornati_al'].astype(str)
```

Adding information from the institutes.

```python

df_info = pd.read_csv('../outputs/clean/istituti_info.csv')
```

```python
df_info = df_info.rename(columns={'id_istituto': 'id'})
```

```python
df_info.head()
```

```python
merged_df = pd.merge(df_most_recent, df_info, on='id')
```

```python
merged_df.head()
```

```python
merged_df.to_csv('../outputs/viz/most_recent.csv', index=False, encoding='UTF-8-sig')

merged_df.to_json('../outputs/viz/most_recent.json', orient='records', lines=True)
```

## Historical

```python
df_historical = df[['id', 'nome', 'tipo', 'posti_regolamentari', 'posti_non_disponibili', 'posti_occupati', 'posti_aggiornati_al']]
df_historical
```

```python
df_historical['posti_reali'] = df_historical['posti_regolamentari'] - df_historical['posti_non_disponibili']
df_historical
```

```python
pd.to_datetime(df_historical['posti_aggiornati_al'], dayfirst=True)
```

```python
# Group by 'posti_aggiornati_al' and aggregate
grouped_df = df_historical.groupby('posti_aggiornati_al').agg({
    'posti_regolamentari': 'sum',
    'posti_non_disponibili': 'sum',
    'posti_occupati': 'sum',
    'posti_reali': 'sum'
}).reset_index()
```

```python
grouped_df
```

```python
grouped_df['tasso_occupazione'] = round(grouped_df['posti_occupati'] / grouped_df['posti_reali'], 2)
grouped_df
```

```python
grouped_df.to_csv('../outputs/viz/historic_data.csv', index=False)
```

```python
grouped_df.to_json('../outputs/viz/historic_data.json', orient='records', lines=True)
```
